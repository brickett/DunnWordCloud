```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
```

```python
text = open(os.path.join(os.getcwd(),'to_director.txt'))
```

```python
IL_mask = np.array(Image.open(os.path.join(os.getcwd(),"IL_2.png")))
```

```python
stopwords = set(STOPWORDS)
stopwords.add("employee")
stopwords.add("employees")
stopwords.add("staff")
stopwords.add("people")
stopwords.add("job")
stopwords.add("work")
stopwords.add("need")
stopwords.add("want")
stopwords.add("agency")
stopwords.add("us")
stopwords.add("good")
stopwords.add("worker")
stopwords.add("state")
stopwords.add("make")
stopwords.add("keep")
stopwords.add("know")
stopwords.add("workers")
stopwords.add("working")
stopwords.add("stop")
```

```python
wc = WordCloud(background_color="white",max_words=2000, mask=IL_mask,stopwords=stopwords)

wc.generate(text.read())

wc.to_file(os.path.join(os.getcwd(),"Q1.png"))

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
